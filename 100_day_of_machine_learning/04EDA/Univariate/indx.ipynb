```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# when we analyze one variable called univariate
df = pd.read_csv('titanic.csv') 
```

```python
df.head(10)
```

```python
# first on categorical data
# COUNTPLOT
# sns.countplot(df['Survived'])
# plt.show()
# or
df['Survived'].value_counts().plot(kind='bar')
```

```python
# pie chart
df['Sex'].value_counts().plot(kind='pie')
```

```python
# numerical data
# histogram is for categorize numerical data between ranges
# like how many item fall in 0 to 10 and 11 to 20 so on 
# histogram
plt.hist(df['Age'],bins=50)
plt.show()
```

```python
    # displot is same as histogram
sns.kdeplot(df['Pclass'])
plt.show()
```

```python
# boxplot this gives us 5 number summary
sns.boxplot(df['Age'])
plt.show()
```
