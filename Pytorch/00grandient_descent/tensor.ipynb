```python
import numpy as np
```

```python
# Create a 5D tensor of shape (2, 3, 4, 5, 6)
tensor_5d = np.random.rand(2, 3, 4, 5, 6)  # Random values for demonstration
```

```python
# Print the shape of the tensor
print("Shape of the 5D Tensor:", tensor_5d)
```

```python
# Access specific elements or slices
# Example: Access the element at position (1, 2, 3, 4, 5)
element = tensor_5d[1, 2, 3, 4, 5]
print("Element at (1, 2, 3, 4, 5):", element)
```

```python
# Slice: Extract a 4D slice by fixing the first index
slice_4d = tensor_5d[1, :, :, :, :]
print("Shape of the 4D slice:", slice_4d)
```
